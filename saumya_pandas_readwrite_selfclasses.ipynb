```python
#loading data
import pandas as pd
url='https://api.github.com/repos/pandas-dev/pandas/issues'
data=pd.read_json(url)
data=pd.DataFrame(data,columns=['number','title','user','state','label'])
data.head(5)
```

```python
#split keys and value in columns
import json
from pandas.io.json import json_normalize
user_data=json_normalize(data['user'])
user_data.head(3)
```

```python
#concatenating data
df3=pd.concat([data[:],user_data[:]],axis=1)

df3.head(5)
```

```python
df4=pd.concat([data[['number','title','state']],user_data[:]],axis=1)
df4.head(5)
```
